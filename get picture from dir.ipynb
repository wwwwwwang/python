```python
import os  
path = "C:/Users/Madhouse/" #文件夹目录  
files= os.listdir(path) #得到文件夹下的所有文件名称
s=[]
for file in files: #遍历文件夹 
    if not os.path.isdir(file): #判断是否是文件夹，不是文件夹才打开
        print(file)
        if(file.find(".jpg") > 0):
            s.append(file) #每个文件的文本存到list中  
    
print(s) #打印结果
```

```python
train_gt="C:/Users/Madhouse/Desktop/wider_face_split/wider_face_split/wider_face_train_bbx_gt1.txt"
with open(train_gt, 'r') as f:
    gts = f.readlines()
count = len(gts)
print("there are %d records in gt file"%count)

str = ""
fw = open('train_info1.txt', 'w')
for gt in gts:
    gt = gt.strip()
    if(gt.find("jpg")>0):
        if(len(str)>0):
            fw.write('%s\n'%(str))
            print("out string = ", str)
            str=gt
            print("one record has been processed!!")
        else:
            str=gt
            continue
    else:
        if(len(gt)<4):
            continue
        else:
            values = gt.split(" ")
            if(int(values[7])==0):
                str = str + " " + values[0] + " " + values[1] + " " + values[2] + " " + values[3]
            else:
                continue
fw.write('%s'%(str))
fw.flush()
fw.close()
print("finished..")
```
