```python
import matplotlib.pyplot as plt  
import numpy as np  
%matplotlib inline
  
  
def sigmoid(x):  
    # 直接返回sigmoid函数  
    return 1. / (1. + np.exp(-x))  
  
  
def plot_sigmoid():  
    # param:起点，终点，间距  
    x = np.arange(-4, 4, 0.1)  
    y = sigmoid(x)
    #plt.plot(x, x)
    plt.plot(x, y)  
    plt.show() 
    
#plot_sigmoid()

def compare():
    x = np.arange(-3.5, 2.5, 0.2)
    y = x*2/(x*2+x-0.5)
    z = sigmoid(x*2/(x-0.5))
    plt.plot(x,y, color="red",  linewidth=2.5, linestyle="-", label="divide")
    plt.plot(x,z, color="g",    linewidth=2.5, linestyle="-", label="sigmoid")
    plt.legend(loc='upper right')
    plt.show() 

print(sigmoid(5/3), 5/8)
compare()

print("0.2:0.3:0.5, sigmoid = (%f:%f:%f)"%(sigmoid(0.2),sigmoid(0.3),sigmoid(0.5)))
sum=sigmoid(0.2)+sigmoid(0.3)+sigmoid(0.5)
print("sigmoid normalization= (%f:%f:%f)"%(sigmoid(0.2)/sum,sigmoid(0.3)/sum,sigmoid(0.5)/sum))

print("2:3:5, sigmoid = (%f:%f:%f)"%(sigmoid(2),sigmoid(3),sigmoid(5)))
sum=sigmoid(2)+sigmoid(3)+sigmoid(5)
print("sigmoid normalization= (%f:%f:%f)"%(sigmoid(2)/sum,sigmoid(3)/sum,sigmoid(5)/sum))
```
