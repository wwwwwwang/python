```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
import random
%matplotlib inline

def getColor():
   r=hex(random.randint(0,255))
   g=hex(random.randint(0,255))
   b=hex(random.randint(0,255))
   return '#'+str(r[2:4]).zfill(2)+str(g[2:4]).zfill(2)+str(b[2:4]).zfill(2)

fig = plt.figure()
ax1 = fig.add_axes([0.05,0.05,0.95,0.95])

#plt.figure(figsize=(16,8))
#m = Basemap()
#m = Basemap(llcrnrlon=73, llcrnrlat=18, urcrnrlon=135, urcrnrlat=53)
#m = Basemap(llcrnrlon=77, llcrnrlat=14, urcrnrlon=140, urcrnrlat=51, projection='lcc', lat_1=33, lat_2=45, lon_0=100)



m = Basemap(llcrnrlon=80.33, 
              llcrnrlat=13.01, 
              urcrnrlon=138.16, 
              urcrnrlat=56.123,
             resolution='h', projection='cass', lat_0 = 42.5,lon_0=120)
             

m.readshapefile('D:/Users/Madhouse/basemap/CHN_adm_shp/CHN_adm1', 'states', drawbounds=True)
m.shadedrelief()
m.drawcoastlines()
m.drawcountries(linewidth=1.5)
m.drawmapboundary()




for info, shp in zip(m.states_info, m.states):
    #proid = info['NAME_1']  # notepadCHN_adm1.csv'NAME_1'
    #if proid == 'Hebei':
    c=getColor()
    poly = Polygon(shp,facecolor=c ,edgecolor='c', lw=3)
    ax1.add_patch(poly)

plt.show()
```
